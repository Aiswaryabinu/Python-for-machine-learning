slicing
syntax : list_name[start : end : step]

    1.start (optional): Index to begin the slice (inclusive). Defaults to 0 if omitted.
    2.end (optional): Index to end the slice (exclusive). Defaults to the length of list if omitted.
    3.step (optional): Step size, specifying the interval between elements. Defaults to 1 if omitted

```python
a=[1,2,3,4,5]
print(a[1:4])
```

Using [:] & [::] without specifying any start, end, or step returns all elements of the list.

```python
a=list(range(0,5))
print(a)
print(a[:])
print(a[::])
print(a[:2])
print(a[2:])
print(a[1:4])
print(a[::2])
print(a[::-1]) #reversing
```

List comprehesion

*   A way to create lists using a concise syntax.
*   It allows us to generate a new list by applying an expression to each item in an existing iterable (such as a list or range).
*  This helps us to write cleaner, more readable code compared to traditional looping techniques.
   





```python
a=[1,2,3,4,5]
res=[item **2 for item in a]
print(res)
```

List comprehension : syntax

    [expression for item in iterable if condition]

for vs list comprehension

for loop requires multiple lines to create a new list by iterating over items and manually adding each one.  
Whereas, list comprehension do the same task in a single line, this makes the code simpler and easier to read.  

```python
#List comprehension with condition
a=[1,2,3,4,5]
va=[res for res in a if res%2==0]
print(va)
```

Dictonary

data structure that stores the value in key: value pairs.

```python
d={1:"apple",2:23,3:"bye"}
print(d)
```

```python
#accessing of elements
print(d[1])
print(d.get(1))
```

```python
#adding and updating
d[4]="cat"
print(d)
d[2]=25
print(d)
```

Removing elements from dict.

del: Removes an item by key.

pop(): Removes an item by key and returns its value.

clear(): Empties the dictionary.

popitem(): Removes and returns the last key-value pair.


```python
print(d)
#using del
del d[2]
print(d)
```

```python
#pop
val=d.pop(3)
print(val)
```

```python
#popitem
key,val=d.popitem()
print(f"Key : {key},value : {val}")
```

```python
#clear
d.clear()
print(d)
```

```python
#iterate through dict
a={1:"apple",2:23,3:"cat",4:"dog",5:45}
print(a)

#key
for key in a:
  print(key)
#value
for value in a.values():
  print(value)
#key,value
for key,value in a.items():
  print(f"{key}:{value}")
```

```python
#nested dict
a={1:12,2:"cat",3:"apple",4:{1:"a",2:"b"}}
print(a)
#length
print(len(a))
```

```python
#check for key
if 1 in a:
  print(a[1])
```

```python
#convert dict to list
b=list(a.items())
print(b)
#traverse
for key in a:
  print(key,a[key])
#copy
c=b.copy()
print(c)
```
